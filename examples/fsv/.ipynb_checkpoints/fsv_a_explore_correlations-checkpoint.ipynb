Load dataset

```julia
dataset_name = "dataset_c"
dataset_a = BSON.load(string(dataset_name,"/",dataset_name,".bson"))
data = dataset_a[:data_tensor].*(-1);
```

Normalise data

```julia
function normalise(d)
    z(s, μ, σ, ϵ) = (s - μ) / (σ + ϵ)
    for i=1:size(d)[2]
        tmp = d[:, i]
        d[:, i] = z.(tmp, mean(tmp), std(tmp), 10e-10)
    end
    return d
end

deflection = normalise(data[:, :, 1]);
loads = normalise(data[:, :, 2]);
```

Show correlations

```julia
#=
plot(xlabel="loads agent 4", ylabel="loads i-th agent")#, xticks=(1, ""), yticks=(1, ""))
scatter!(loads[:, 4], loads[:, 5], label="a5")
scatter!(loads[:, 4], loads[:, 20], label="a20")
scatter!(loads[:, 4], loads[:, 50], label="a50")
=#
```

Detect outliers - function

```julia
function corroutliers(df::DataFrame, a::Integer, b::Integer)
    reg = createRegressionSetting(@formula(y~x), df)
    outliers = ccf(reg)["outliers"]
    layout = Layout(;xaxis=attr(title=string("point ", a)),
                     yaxis=attr(title=string("point ", b)))
    p1=scatter(;x=df[findall(x -> x ∉ outliers, 1:size(df)[1]), :x], 
              y=df[findall(x -> x ∉ outliers, 1:size(df)[1]), :y], 
             marker=attr(color="rgba(20, 20, 255, 0.95)"), mode="markers", showlegend=false)
    p2=scatter(;x=df[findall(x -> x ∈ outliers, 1:size(df)[1]), :x], 
             y=df[findall(x -> x ∈ outliers, 1:size(df)[1]), :y], 
             marker=attr(color="rgba(255, 20, 20, 0.95)"), mode="markers", showlegend=false)
    return plot([p1, p2], layout)
end
```

```julia
p1 = corroutliers(DataFrame(y=loads[:, 18], x=loads[:, 20]), 18, 20)
p2 = corroutliers(DataFrame(y=loads[:, 30], x=loads[:, 31]), 30, 31)
p3 = corroutliers(DataFrame(y=loads[:, 44], x=loads[:, 45]), 44, 45)
p4 = corroutliers(DataFrame(y=loads[:, 10], x=loads[:, 30]), 10, 30)
p=[p1 p2; p3 p4]
savefig(p, string("/Users/drvojtex/Desktop/", dataset_name, "_loads_points.pdf"))
p
```
