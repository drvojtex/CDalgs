# CDalgs - correlations explorations on fsv dataset

There are examples of correlations between agents features.

Load libraries

```julia
using BSON, DataFrames
using LinRegOutliers, StatsBase
using Plots
```

Load dataset

```julia
dataset_a = BSON.load("dataset_a.bson")
data = dataset_a[:data_tensor].*(-1);
```

Normalise data

```julia
function normalise(d)
    z(s, μ, σ, ϵ) = (s - μ) / (σ + ϵ)
    for i=1:size(d)[2]
        tmp = d[:, i]
        d[:, i] = z.(tmp, mean(tmp), std(tmp), 10e-10)
    end
    return d
end

deflection = normalise(data[:, :, 1]);
loads = normalise(data[:, :, 2]);
```

Show correlations

```julia
plot(xlabel="loads agent 4", ylabel="loads i-th agent")#, xticks=(1, ""), yticks=(1, ""))
scatter!(loads[:, 4], loads[:, 5], label="a5")
scatter!(loads[:, 4], loads[:, 20], label="a20")
scatter!(loads[:, 4], loads[:, 50], label="a50")
```

Detect outliers - function

```julia
function corroutliers(df::DataFrame, a::Integer, b::Integer)
    reg = createRegressionSetting(@formula(y~x), df)
    outliers = ccf(reg)["outliers"]
    p = plot(xlabel = string("agent ", a), 
             ylabel = string("agent ", b))
    scatter!(df[findall(x -> x ∉ outliers, 1:size(df)[1]), :x], 
             df[findall(x -> x ∉ outliers, 1:size(df)[1]), :y], 
             label = false)
    scatter!(df[findall(x -> x ∈ outliers, 1:size(df)[1]), :x], 
             df[findall(x -> x ∈ outliers, 1:size(df)[1]), :y], 
             color=:red, label = "outliers")
    return p
end
```

```julia
p1 = corroutliers(DataFrame(y=loads[:, 18], x=loads[:, 20]), 18, 20)
p2 = corroutliers(DataFrame(y=loads[:, 30], x=loads[:, 31]), 30, 31)
p3 = corroutliers(DataFrame(y=loads[:, 44], x=loads[:, 45]), 44, 45)
p4 = corroutliers(DataFrame(y=loads[:, 10], x=loads[:, 30]), 10, 30)
p = plot(p1, p2, p3, p4, layout=(2,2))#, xticks=(1, ""), yticks=(1, ""))
p[:plot_title] = "Loads"
plot(p)
```

```julia
p1 = corroutliers(DataFrame(y=deflection[:, 18], x=deflection[:, 20]), 18, 20)
p2 = corroutliers(DataFrame(y=deflection[:, 30], x=deflection[:, 31]), 30, 31)
p3 = corroutliers(DataFrame(y=deflection[:, 44], x=deflection[:, 45]), 44, 45)
p4 = corroutliers(DataFrame(y=deflection[:, 10], x=deflection[:, 30]), 10, 30)
p = plot(p1, p2, p3, p4, layout=(2,2))#, xticks=(1, ""), yticks=(1, ""))
p[:plot_title] = "Deflection"
plot(p)
```
